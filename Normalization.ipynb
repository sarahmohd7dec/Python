```python
import pandas as pd
#feature Scaling-range is from 0 to 1
df = pd.DataFrame({'Height':[6,3,4,6.2,8],
                     
                   'Weight':[75,40,56,200,58]})
  
display(df)
df['Weight']=df['Weight']/df['Weight'].max()
df
```

```python
#Z-score range is from -3 to +3
df = pd.DataFrame({'Height':[6,3,4,6.2,8],
                     
                   'Weight':[75,40,56,200,58]})
  
display(df)
df['Weight']=(df['Weight']-df['Weight'].mean())/(df['Weight'].std())
df
```

```python
#Min-Max:range is from 0 to 1
df = pd.DataFrame({'Height':[6,3,4,6.2,8],
                     
                   'Weight':[75,40,56,200,58]})
  
display(df)
df["Weight"]=(df["Weight"]-df["Weight"].min())/(df["Weight"].max()-df["Weight"].min())
df
```
