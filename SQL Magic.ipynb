```python
 #How many Elementary Schools are in the dataset?
%sql Select * from Chicago_Public_Schools
%sql select count(ElementaryMiddleorHighSchool)as NumberofElemntarySchools from Chicago_Public_Schools where ElementaryMiddleorHighSchool='ES' group by ElementaryMiddleorHighSchool 
```

```python
# Display total number of elementary, middle and high school from Chicago_public_Schools
%sql Select ElementaryMiddleorHighSchool,count(ElementaryMiddleorHighSchool)as NumberofSchools from Chicago_Public_Schools group by ElementaryMiddleorHighSchool
```

```python
#What is the highest Safety Score? #Which schools have highest Safety Score
%sql select NAME_OF_SCHOOL from Chicago_Public_Schools where SAFETY_SCORE=(SELECT max(SAFETY_SCORE)from Chicago_Public_Schools) group by SAFETY_SCORE
```

```python
#Remove the '%' sign for Average Student Attendance column
%sql Select REPLACE(AVERAGE_STUDENT_ATTENDANCE,'%','')as Remove from Chicago_Public_Schools
```

```python
#Which Schools have Average Student Attendance lower than 70%?
%sql SELECT NAME_OF_SCHOOL from Chicago_Public_Schools where AVERAGE_STUDENT_ATTENDANCE<70
```

```python
#Get the total College Enrollment for each Community Area.
%sql SELECT COMMUNITY_AREA_NAME,count(COLLEGE_ENROLLMENT)as COUNT from Chicago_Public_Schools group by COMMUNITY_AREA_NAME
```

```python
# Get the 5 Community Areas with the least College Enrollment sorted in ascending order
%sql SELECT COMMUNITY_AREA_NAME,count(COLLEGE_ENROLLMENT)as COUNT from Chicago_Public_Schools group by COMMUNITY_AREA_NAME order by count(COLLEGE_ENROLLMENT)LIMIT 5
```

```python
#Display ZIPCodes from Chicago_Schools which are having more than 5 schools

%sql SELECT ZIPCode,count(ZIPCode)as Count from Chicago_Public_Schools group by ZIPCode having count(ZIPCode)>5 order by COUNT DESC
```

```python
 #Display the type of crimes with less than 10 reported crimes for each category.
%sql SELECT PRIMARY_TYPE as typeofcrime,count(PRIMARY_TYPE)as count from Chicago_Crime_Data group by PRIMARY_TYPE having count(PRIMARY_TYPE)<10 order by count desc
```

```python
#: Display the top 20 least reported types of crime
%sql SELECT PRIMARY_TYPE as typeofcrime,count(PRIMARY_TYPE)as count from Chicago_Crime_Data group by PRIMARY_TYPE order by count LIMIT 20
```
