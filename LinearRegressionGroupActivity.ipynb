```python
#Read the data from the provided link
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/Kushal997-das/THE-SPARKS-FOUNDATION/master/Prediction%20using%20Supervised%20ML/student_scores%20-%20student_scores.csv")
data.head(5)
```

```python
#Understand the data by reading first 5 and last 5 records from the dataset
print(data.head(5))
print(data.tail(5))
```

```python
#Check the shape of data
data.shape
```

```python
#Check if there is any null values
data.isnull().sum()
```

```python
#Display the summary of the dataset
data.describe()
```

```python
#Group the data using score and hours, then Using bar chart ,plot score vs hours
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
grouped_data = data.groupby(['Scores', 'Hours']).size().reset_index(name='Count')
plt.figure(figsize=(10, 6))
sns.barplot(x='Scores', y='Hours', data=grouped_data)
plt.title('Score vs Hours')
plt.xlabel('Scores')
plt.ylabel('Hours')
plt.show()
```

```python
#Visualize distribution of data using scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Hours', y='Scores', data=data)
plt.title('Hours Studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.show()
```

```python
#implement linear regression
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
x=data[['Hours']]
y=data[['Scores']]
lm.fit(x,y)
yhat=lm.predict(x)
print(yhat)
sns.regplot(x='Hours',y='Scores',data=data)
plt.ylim(0,)
```

```python
#implement polynomial regression


f=np.polyfit(x,y,3)
p=np.poly1d(f)
print(p)
```

```python
from sklearn.preprocessing import PolynomialFeatures
 
poly = PolynomialFeatures(degree=4)
x=x.values.reshape(-1,1)
X_poly = poly.fit_transform(x)
 
poly.fit(X_poly, y)
lin2 = LinearRegression()
lin2.fit(X_poly, y)
```

```python
# Visualising the Polynomial Regression results
plt.scatter(x, y, color='blue')
 
plt.plot(x, lin2.predict(poly.fit_transform(x)),
         color='red')
plt.title('Polynomial Regression')
plt.xlabel('Hours')
plt.ylabel('Scores')
 
plt.show()
```
